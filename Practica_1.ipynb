```python
n = np.arange(0,101, 0.1)
y = np.sin(np.pi * 0.12 * n)
y2 = np.cos(2 * np.pi * 0.03 * n)
s = y + y2
t = y * y2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(n, y, "-r", label = "Sen(π * 0.12n)")
plt.plot(n, y2, "-b", label = "Cos(2π * 0.03n)")
plt.legend()
plt.title("Funcion Y1 y Y2")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(n, s, "-g", label = "Suma")
plt.plot(n, t, "-m", label = "Producto")
plt.legend()
plt.title("Operacione de señales")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.show()
```

```python
notas_curso = {
    "Ana García": 9.5,
    "Luis Rodríguez": 7.0,
    "Marta Sánchez": 8.2,
    "Juan Pérez": 6.4,
    "Sofía Castro": 10.0
}
def notas_seleccionadas(notas):
  notas=np.array(list(notas.values()), dtype=float)
  maxima=np.max(notas)
  minimo=np.min(notas)
  media=np.mean(notas)
  desviacion=np.std(notas)
  serie = pd.Series ([maxima,minimo,media,desviacion], index = ["Nota maxima", "Nota minima", "Media de notas", "Desviación estandar"], name = "Resultados")
  return serie
notas_seleccionadas(notas_curso)
```

```python
carga = pd.read_csv("pacientes_VitalDB (1).csv")
print("--------------------------------------------------------------------------------")
print(carga.head(3))
print("--------------------------------------------------------------------------------")
print(carga.tail(3))
print("--------------------------------------------------------")
print(carga.info())
print("--------------------------------------------------------")
carga.rename(columns = {"height": "weight", "weight": "height"}, inplace = True)
print(carga.info())
print("--------------------------------------------------------")
carga["BMI"] = carga["weight"]  / ((carga["height"]/100) ** 2)
print(carga.info())
print("--------------------------------------------------------")
print(carga["BMI"])
print("--------------------------------------------------------")
limites = [0, 18.5, 25.0, 30.0, float('inf')]
etiquetas = ['Bajo peso', 'Normal', 'Sobrepeso', 'Obesidad']
carga["Categoria_IMC"] = pd.cut(carga["BMI"], bins=limites, labels=etiquetas, right=False)
print(carga[["BMI", "Categoria_IMC"]])
```
