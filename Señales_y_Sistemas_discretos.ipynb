```python
import numpy as np
import matplotlib.pyplot as plt
```

#Impulso

```python
def impulso (n0,n1,n2):
  n=np.arange(n1,n2+1)
  x= (n==n0).astype(int)
  return n,x

x,y=impulso(4,-2,8)
plt.figure(figsize=(10,5))
plt.stem(x,y)
plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Escalón unitario")
plt.grid()
plt.show()
```

#Escalón Unitario

```python
def escalon (n0,n1,n2):
  n=np.arange(n1,n2+1)
  x= (n>=n0).astype(int)
  return n,x

x,y=escalon(4,-2,8)
plt.figure(figsize=(10,5))
plt.stem(x,y)
plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Escalón unitario")
plt.grid()
plt.show()
```

#Funcion rampa

```python
def rampa (n0,n1,n2):
  n3 = n0+1
  n=np.arange(n1,n2+1)
  x= (n>=n3).astype(int)
  cont=0
  for i in x:
    cont+=1
    if i==1:
      x[cont-1]=x[cont-2]+1
  return n,x

x,y=rampa(0,-10,10)
plt.figure(figsize=(10,5))
plt.stem(x,y)
print(type(y))
plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Rampa")
plt.grid()
plt.show()
```

#Exponencial compleja

```python
def exp_complex (n0,n1,n2,sigma,w0):
  n=np.arange(n1,n2+0.5)
  arg = complex(sigma,w0)
  x=np.exp(arg*n)
  return n,x

x,y=exp_complex(0.1,-10,10,0.05,0.2 * np.pi)
plt.figure(figsize=(10,8))
plt.stem(x,y)
plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Exponencial compleja")
plt.grid()
plt.show()
```

# Sinusoides

```python
def sinusoide (n0,n1,n2,w0,desfase):
  n = np.arange(n1,n2+n0,n0)
  x = np.sin(w0*n + desfase)
  return n,x

x,y=sinusoide(0.1,-10,10,1,0)
plt.figure(figsize=(10,5))
plt.stem(x,y)
print(type(y))
plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Sinusoide")
plt.grid()
plt.show()



```

#Punto 6

```python
n = np.arange(0, 18)
x = np.zeros(len(n))

for i, n1 in enumerate(n):
    if 0 <= n1 <= 5:
        x[i] = n1
    elif 6 <= n1 <= 11:
        x[i] = n1 - 5
    elif 12 <= n1 <= 17:
        x[i] = n1 - 10

dx = np.diff(x, prepend=0)

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.stem(n, x, linefmt='b-', markerfmt='bo')
plt.title('Señal Original X[n]')
plt.grid()

plt.subplot(2, 1, 2)
plt.stem(n, dx, linefmt='r-', markerfmt='ro')
plt.title('Y[n] = X[n] - X[n-1]')
plt.xlabel('n')
plt.grid()
plt.tight_layout()
plt.show()
```

# Punto 7

```python
def escalon(n, n0):
    return np.where(n >= n0, 1, 0)

def rampa(n, n0):
    return np.where(n >= n0, n - n0, 0)

def x(n):
    r1 = rampa(n, 0)
    r2 = rampa(n, 5)
    r3 = rampa(n, 10)
    r4 = rampa(n, 15)
    u1 = escalon(n, 19)
    return r1 - 2*r2 + 2*r3 - r4 + 5*u1

n=np.arange(-10,25)
x_n=x(n)

plt.figure(figsize=(10, 5))
plt.stem(n, x_n)
plt.title('Secuencia')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid()
plt.show()

```

```python
# a. x_5(n) = 2x(n - 4) + x(n)

n_a = np.arange(-5, 35)
x_5 = 2 * x(n_a - 4) + x(n_a)

plt.figure(figsize=(10, 5))
plt.stem(n_a, x_5)
plt.title('Secuencia a')
plt.xlabel('n')
plt.ylabel('Valor')
plt.grid()
plt.show()

# b. x_6(n) = 0.001 * e^(0.5n) * x(n) + 10 * sin(0.05 * pi * n) * x(n + 2)

n_b = np.arange(-20, 21)
x_6 = 0.001 * np.exp(0.5 * n_b) * x(n_b) + 10 * np.sin(0.05 * np.pi * n_b) * x(n_b + 2)

plt.figure(figsize=(10, 5))
plt.stem(n_b, x_6)
plt.title('Secuencia b')
plt.xlabel('n')
plt.ylabel('Valor')
plt.grid()
plt.show()
```
